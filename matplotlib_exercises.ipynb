```python
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
# 1. Use matplotlib to plot the following equation:
# y = x^2 − x + 2
# You'll need to write the code that generates the x and y points.
# Add an anotation for the point 0, 0, the origin.

x = range(-5, 10)
y = [((n**2)-n+2) for n in x]

plt.scatter(x, y)
plt.annotate('Origin', xy=(0, 0), xytext=(1, 10), arrowprops={'facecolor':'blue'})
plt.xlabel('x')
plt.ylabel('$x^2 - x + 2$')
plt.title('$1. y = x^2 - x + 2$')
plt.axis([-6, 10, 0, 80])
plt.show()
```

```python
# 2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):
# y = √x

x1 = range(0,41)
y1 = [math.sqrt(n) for n in x1]

plt.scatter(x1,y1)
plt.xlabel('x')
plt.ylabel('√x')
plt.title('2a. y = √x') 
plt.axis([0,41,0,7])
```

```python
# y = x^3
x2 = range(0,4)
y2 = [n**3 for n in x2]

plt.scatter(x2,y2)
plt.xlabel('x')
plt.ylabel('$x^3$')
plt.title('$2b. y = x^3$')
plt.axis([0,11,0,30])
```

```python
# y = 2^x
x3 = range(0, 5)
y3 = [2**n for n in x3]

plt.scatter(x3,y3)
plt.xlabel('x')
plt.ylabel('$2^x$')
plt.title('$2c. y = 2^x$')
plt.axis([0,11,0,20])
```

```python
# y = 1/(x + 1)
x4 = range(0,20)
y4 = [(1/(n+1)) for n in x4]

plt.scatter(x4,y4)
plt.xlabel('x')
plt.ylabel('$1/(x+1)$')
plt.title('2d. y = 1/(x+2)')
plt.axis([-.5,20,0,1.1])
```

```python
# 3. Combine the figures you created in the last step into one large figure with 4 subplots.
n_rows = 2
n_cols = 2

plt.subplot(n_rows, n_cols, 1)
plt.scatter(x1, y1)
plt.title('y = √x')

plt.subplot(n_rows, n_cols, 2)
plt.scatter(x2, y2)
plt.title('$y = x^3$')

plt.subplot(n_rows, n_cols, 3)
plt.scatter(x3, y3)
plt.title('$y = 2^x$')

plt.subplot(n_rows, n_cols, 4)
plt.scatter(x4, y4)
plt.title('y = 1/(x+2)')

plt.subplots_adjust(top=1,bottom=0)
plt.suptitle('Question 2', y=1.10)
```

```python
# 4. Combine the figures you created in the last step into one figure where each of the 4 equations 
# has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

plt.scatter(x1, y1, c='blue', label='√x')

plt.scatter(x2, y2, c='red', label='$x^3$')

plt.scatter(x3, y3, c='orange', label='$2^x$')

plt.scatter(x4, y4, c='yellow', label='1/(x+2)')
plt.axis([0,45,-1,30])
plt.legend(loc='upper right')

plt.show()
```
